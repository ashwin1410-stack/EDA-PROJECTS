# **Top 5 most popular artists**

```python
five_most_popular_artists = data.groupby("artist").count().sort_values(by="song_title",ascending = False)["song_title"][:5]
five_most_popular_artists
```

```python
five_most_popular_artists.plot.barh()
plt.subplot().set_title("five_most_popular_artists")
plt.show()
```

**Top 5 most loudest tracks**

```python
five_most_loudest_track = data[["loudness","song_title"]].sort_values(by="loudness",ascending = True)[:5]
five_most_loudest_track
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x="loudness",y="song_title",data=five_most_loudest_track)
plt.title("Top 5 most Loudest tracks")
plt.show()
```

**Artists with most danceability song**

```python
top_five_artist_with_most_danceability_songs = data[["danceability","song_title","artist"]].sort_values(by="danceability",ascending = False)[:5]
top_five_artist_with_most_danceability_songs
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x="danceability",y="artist",data=top_five_artist_with_most_danceability_songs)
plt.title("Artist with the most danceability song")
plt.show()
```

**Top 10 instrumentals tracks**

```python
top_ten_instrumentals_tracks = data[["instrumentalness","song_title","artist"]].sort_values(by="instrumentalness",ascending=False)[:10]
top_ten_instrumentals_tracks
```

```python
plt.figure(figsize=(12,7))
plt.pie(x="instrumentalness",data=top_ten_instrumentals_tracks,autopct="%1.2f%%",labels = top_ten_instrumentals_tracks.song_title)
plt.title("Top 10 instrumentalness")
plt.show()
```

**Multiple Feature Plots**

```python
data.columns
```

```python
interests_feature_col = ["tempo","loudness","acousticness","danceability","duration_ms","energy","instrumentalness","liveness","speechiness","valence"]
```

```python
for col in interests_feature_col:
  pos_data = data[data["target"]==1][col]
  neg_data = data[data["target"]==0][col]

  plt.figure(figsize=(12,7))
  sns.distplot(pos_data,bins=30,label= "Positive",color="blue")
  sns.distplot(neg_data,bins=30,label="Negative",color="green")
  plt.legend(loc="upper right")
  plt.title(f"Positive & Negative histogram plots for {col}")
  plt.show

```

*Finding co-realtion between all the value*

```python
dataplot = sns.heatmap(data.corr(),cmap="YlGnBu",annot=True)
sns.set(rc = {"figure.figsize":(20,10)})
plt.show()
```

*What do we find out by above figure*

*  Skills moves has Negative correlation with **Height & Weight** .It means more heighted and more weighted the player has lesser skills he will have
* Age has little bit positive corealtion with WAGE
*Whereas AGE has strong NEGATIVE CORRELATION with POTENTIAL but POSITIVE CORREALTON with overall rating
* OVERALL RATING has Strong POSITIVE CORREALTION with RAGE & REALEASE CLAUSE

**Highest Wages Player**

```python
data_max_wages=data.sort_values(by="Wage",ascending=0).head(10)
fig,ax= plt.subplots()
ax1=sns.barplot(data=data_max_wages,x="Name",y="Wage")
ax2 =sns.catplot(data=data_max_wages,x="Position",y="Wage")
ax3= sns.catplot(data=data_max_wages,x="Club",y="Wage")
plt.xticks(rotation=90,fontsize=18)
```

**Highest Overall Player**

```python
data_max_overall=data.sort_values(by="Overall",ascending=0).head(10)
ax1=sns.catplot(data=data_max_overall,x="Name",y="Overall",label="Club")
plt.xticks(rotation=90,fontsize=8)
ax2=sns.catplot(data=data_max_overall,x="Position",y="Overall")
ax3= sns.catplot(data=data_max_overall,x="Club",y="Overall")
plt.xticks(rotation=90,fontsize=8)
```

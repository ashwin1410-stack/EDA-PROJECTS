# ***EDA***

*** GENDER ***

```python
df.columns
```

```python
ax = sns.countplot(x = 'Gender', data = df)
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'] , as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
```

```python
sns.barplot( x = 'Gender' ,y = 'Amount' ,data = sales_gen) 
```

From above  graph we can see that most of the buyers is females and purchasing power of female  is greater than male 

```python
ax = sns.countplot(x = 'Age Group', data = df, hue = 'Gender')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
#Toltal amount vs Age group
sales_age = df.groupby(['Age Group'] , as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
```

```python
sns.barplot(x = 'Age Group', y = 'Amount' , data = sales_age)
```

**FROM ABOVE GRAPH WE HAVE OBSERVE THAT MOST OF THE BUYERS ARE AGE GROUP BETWWN 26-35** 

# STATE

```python
df.columns
```

```python
sales_state = df.groupby(['State'] , as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)
```

**Total number of order from top 10 state**

```python
sns.set(rc={'figure.figsize': (20,10)})
sns.barplot(data = sales_state ,y ='Orders', x = 'State')
```

**Total amount/sales from top 10 state**

```python
sales_state = df.groupby(['State'] , as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
```

```python
sns.set(rc={'figure.figsize': (20,10)})
sns.barplot(data = sales_state ,y ='Amount', x = 'State')
```

From above graph we can  see that most of the orders are from th Uttar pradesh,Maharashtra and Karnataka 

**MARTIAL STATUS**

```python
ax = sns.countplot(data =df , x ='Marital_Status')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'] , as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
```

```python
sns.set(rc={'figure.figsize': (6,5)})
sns.barplot(data = sales_state , x = 'Marital_Status' , y = 'Amount', hue = 'Gender')
```

From the above graph we can see that most of the buyers is married women and they have high purchasing power

**OCCUPATION**

```python
sns.set(rc={'figure.figsize': (20,16)})
ax = sns.countplot(data =df , x ='Occupation')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'] , as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.set(rc={'figure.figsize': (20,18)})
sns.barplot(data = sales_state , x = 'Occupation' , y = 'Amount')
```

From the above graph we cans see that most of the buyers are in ITSECTOR , HEALTHCARE AND AVIATION

**PRODUCT_CATEGORY**

```python
sns.set(rc={'figure.figsize': (28,5)})
ax = sns.countplot(data =df , x ='Product_Category')
for bars in ax.containers:
    ax.bar_label(bars)
```

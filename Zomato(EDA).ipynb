*Reading EXCEL Data*

```python
data_country=pd.read_excel("/content/Country-Code.xlsx")
data_country.head()
```

*Combining the datasets*

```python
final_data=pd.merge(data,data_country,how ='left',on='Country Code')
final_data
```

Checking Dtypes

```python
final_data.info()
```

Checking Columns

```python
final_data.columns
```

```python
final_data.Country.value_counts()
```

* **Observation :-**

    * From above, we get that there is lot of Zomato hub in INDIA & very less in CANADA & USA stand 2nd in the list by 434 Hub or Startup

```python
country_name=final_data.Country.value_counts().index
print(country_name)
country_value=final_data.Country.value_counts().values
print(country_value)
```

*Top 5 countries Uses Zomato*

```python
plt.figure(figsize=(10,12))
plt.pie(country_value[:5],labels=country_name[:5],autopct='%.2f')
plt.title("Top 5 Countries that uses Zomato")
plt.legend()
plt.show()
```

*  **Observation :-**
  
    *  Zomato maximun records OR transaction is from INDIA
    *  After that there is USA & United Kingdom

```python
rating= final_data.groupby(["Aggregate rating","Rating color","Rating text"]).size().reset_index().rename(columns={0:'Rating Count'})
rating
```

*   **Observation :-**
  *  When the rating is between 4.5 to 4.9 it is ----> *Excellent*
  *  When the rating is between 4.0 to 4.4 it is ----> *Very Good*
  *  When the rating is between 3.5 to 3.9 it is ----> *Good*
  *  When the rating is between 2.5 to 3.4 it is ----> *Average*
  *  When the rating is between 1.8 to 2.4 it is ----> *Poor*

```python
rating.head()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x="Aggregate rating",y="Rating Count",data=rating,hue='Rating color',palette=['blue','red','orange','yellow','green','green'])
plt.xlabel("Aggregrate Ratings")
plt.ylabel("Count")
plt.title("Aggregate Rating V/S  Rating Counts")
plt.show()
```

*  **Observation:-**

  *  Not Rated Count is  very high
  *  Maximum numbers of rating are between 2.5 to 3.4

```python
plt.figure(figsize=(10,6))
sns.countplot(x="Rating color",data=rating,palette=['blue','red','orange','yellow','green','green'])
plt.xlabel("Ratings color")
plt.ylabel("Count")
plt.title("Rating color  V/S   Counts")
plt.show()
```

*  **Observation**
    
    *  Orange is very more ratings rather than all of them
    *  White is very less i.e only 1 rating
    *  Yellow, Green & Dark green is eqaully counts
